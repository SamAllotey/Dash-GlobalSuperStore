```python
# file=os.path.join("globalSuperstore","newSuperstore.csv")
# xlsdf.to_csv(file,index=False)
```

```python
xlsdf['Profit_Margin'] = (xlsdf['Profit']/ xlsdf['Sales'])*100
```

```python
xlsdf.head(7)
```

```python
#Most Profitable Category
best_selling_pivot = pd.pivot_table(xlsdf[['Category','Sales','Profit_Margin','Product_Name']],
                                    index=['Category','Profit_Margin','Product_Name'], aggfunc='sum')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by = ['Profit_Margin'],ascending=False).head().sort_values(by='Sales',ascending=False)
```

```python
#Most Profitable Sub-Category
best_selling_pivot = pd.pivot_table(xlsdf[['Sub_Category','Sales','Profit_Margin','Product_Name']],
                                    index=['Sub_Category','Profit_Margin','Product_Name'], aggfunc='sum')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by = ['Profit_Margin'],ascending=False).head().sort_values(by='Sales',ascending=False)
```

```python
xlsdf['Sales_Count'] = 1
```

```python
#Top Selling Sub-Category
best_selling_pivot = pd.pivot_table(xlsdf[['Sub_Category','Sales_Count','Product_Name']],
                                    index=['Sub_Category','Product_Name'], aggfunc='count')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by=['Sales_Count'],ascending=False).head(15)
```

```python
#Which customer segment is most Profitable
best_selling_pivot = pd.pivot_table(xlsdf[['Segment','Sales','Profit_Margin']],
                                    index=['Segment','Profit_Margin'], aggfunc='count')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by=['Profit_Margin'],
                               ascending=False).head(15).sort_values(by=['Sales'],ascending=False)
```

```python
#preferred ship mode
best_selling_pivot = pd.pivot_table(xlsdf[['Ship_Mode','Sales_Count']],
                                    index=['Ship_Mode'], aggfunc='count')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by=['Sales_Count'],ascending=False).head()
```

```python
 xlsdf['Ship_Mode'].value_counts()
```

```python
#most profitable region
best_selling_pivot = pd.pivot_table(xlsdf[['Region','Sales','Profit_Margin']],
                                    index=['Region','Profit_Margin'], aggfunc='sum')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by=['Profit_Margin'],
                               ascending=False).head(15).sort_values(by=['Sales'],ascending=False).head(5)
```

```python
#which city has the highest sales
best_selling_pivot = pd.pivot_table(xlsdf[['City','Sales_Count']],
                                    index=['City'], aggfunc='count')

best_selling_pivot=best_selling_pivot.reset_index()

best_selling_pivot.sort_values(by=['Sales_Count'],
                               ascending=False).head()
```
